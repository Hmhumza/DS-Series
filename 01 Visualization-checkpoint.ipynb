```python
#for making bar plot for categorical variable

import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks",color_codes=True)


titanic = sns.load_dataset("titanic")
sns.catplot(x="sex",y="survived",hue="class",kind="bar",data=titanic)
plt.show()
```

```python
#for making count-plot for categorical variable

import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks",color_codes=True)


titanic = sns.load_dataset("titanic")
p1=sns.countplot(x="sex",data=titanic,hue="class")
p1.set_title("plot for counting")
plt.show()
```

```python
#for making scattered for continous variable

import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks",color_codes=True)


titanic = sns.load_dataset("titanic")
g=sns.FacetGrid(titanic,row="sex",hue="alone")
g=(g.map(plt.scatter,"age","fare").add_legend())
plt.show()
```

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
#load daatset
phool = sns.load_dataset("iris")
phool

#draw a lineplot
sns.lineplot(x="sepal_length",y="sepal_width",data=phool)
sns.set_style("ticks")
sns.set_style(style=None,rc=None)
plt.show
```

## Size of figure

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
#load daatset
phool = sns.load_dataset("iris")
plt.figure(figsize=(8,6))

#draw a lineplot
sns.lineplot(x="sepal_length",y="sepal_width",data=phool)

sns.set_style("dark")
sns.set_style(style=None,rc=None)
plt.show
```

# 02 Bar Plot

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
#load daatset
kashti = sns.load_dataset("titanic")
kashti
plt.figure(figsize=(8,6))

#draw a lineplot
sns.barplot(x="sex",y="alone",hue="who",data=kashti,order=["female","male"],color="grey")
sns.set_style("ticks")
sns.set_style(style=None,rc=None)
plt.show
```

```python
#remove error bar by
## ci=None (confidence interval)
#3 by using color palete u can change color ==> palette=pastel
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
#load daatset
kashti = sns.load_dataset("titanic")
kashti
plt.figure(figsize=(8,6))

#draw a lineplot
sns.barplot(x="sex",y="alone",hue="who",data=kashti,order=["female","male"],color="red",ci=None,saturation=0.3)
sns.set_style("ticks")
sns.set_style(style=None,rc=None)
plt.show
```

```python
# facecolor(R, G, B, Alpha) and RGBA values should be within the 0 to 1 range to make different colors.
#Alpha means opacity of the facecolor and its value also should be in the 0 to 1 range.

import seaborn as sns
import matplotlib.pyplot as plt
#load daatset
kashti = sns.load_dataset("titanic")
kashti

sns.barplot(x="class",y="fare",data=kashti,
            linewidth=3, facecolor=(0.1,0.3,0.3,0.5), errcolor="0.5", edgecolor="0")

plt.show
```
